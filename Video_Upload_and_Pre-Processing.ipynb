```python
import cv2
import ffmpeg
import os

def extract_audio(video_path, audio_path):
    (
        ffmpeg
        .input(video_path)
        .output(audio_path, format='wav', acodec='pcm_s16le', ac=1, ar='16000')
        .overwrite_output()
        .run()
    )

def extract_frames(video_path, frames_dir, fps=1):
    os.makedirs(frames_dir, exist_ok=True)
    vidcap = cv2.VideoCapture(video_path)
    count = 0
    while True:
        success, image = vidcap.read()
        if not success:
            break
        if count % int(vidcap.get(cv2.CAP_PROP_FPS)) == 0:
            frame_path = os.path.join(frames_dir, f"frame_{count}.jpg")
            cv2.imwrite(frame_path, image)
        count += 1
    vidcap.release()

video_file = 'video.mp4'
audio_file = 'audio.txt'
frames_directory = 'path/to/frames/'

extract_audio(video_file, audio_file)
extract_frames(video_file, frames_directory)
```
